```python
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
import time

from IPython.display import Image
import torch
from collections import defaultdict
import imageio
import numpy as np
from tqdm import tqdm

import torch
import torch.nn as nn
import random

import math

from get_tensor import get_samples, get_batch, display_image
```

```python
csv_src = "other_data\coords-scattered-hdr.csv"
skies_data = pd.read_csv(csv_src)
skies_data.insert(0, "TensorIndex", [num for num in range(len(skies_data))], False)
skies_data[skies_data.columns[:21]]
```

```python
spectrum_beginning = 350
spectrum_end = 1780
n_frequencies = 1780 - 350 + 1

y = skies_data[
    [
        str(number)
        for number in range(spectrum_beginning, spectrum_end + 1)
    ]
]

y_mean = y.mean(0)
y_range = y.max(0) - y.min(0)

normalized_y = (y - y_mean) / y_range # maybe use range instead of std
normalized_y_tensor = torch.tensor(normalized_y.values).float()
normalized_y_tensor
```

```python
# normalization
# we divide the brightness by 255 in the get_samples code
load_samples = True
save_location = 'other_data\sky_samples_not_random'
if load_samples:
    sky_sample_tensors = torch.load('other_data\sky_samples_not_random')
else:
    sky_sample_tensors = get_batch(skies_data, 0, len(skies_data), get_samples)
    torch.save(sky_sample_tensors, save_location)

#sky_sample_tensors.shape, sys.getsizeof(sky_sample_tensors.storage())

print(normalized_y_tensor[0])
print(sky_sample_tensors[0,23, 48:-48, 48:-48])
```

```python
# These skies are the ones we will use to validate our model against the original model
demo_skies = [ # first 4 clear, next 4 scattered, last 3 overcast
    ("5/26/2013", "15:15"), # (date, time)
    ("5/27/2013", "10:15"),
    ("7/26/2013", "13:15"),
    ("9/24/2013", "15:39"),
    ("5/12/2013", "13:00"),
    ("5/26/2013", "12:30"),
    ("7/29/2013", "10:00"),
    ("9/26/2013", "11:50"),
    ("4/14/2013", "11:36"),
    ("4/15/2013", "08:40"),
    ("7/29/2013", "13:30")
]
date_to_time = {sky[0]: sky[1] for sky in demo_skies}

drop_indexes = []

for index, series in skies_data.iterrows():
    if series["Date"] in date_to_time:
        if series["Time"][:5] == date_to_time[series["Date"]]:
            print(index, list(series[["Date", "Time", "SamplePatternIndex"]]))
            drop_indexes.append(index)
pruned_skies_data = skies_data.drop(index=drop_indexes).reset_index(drop=True)

prev_date_time = tuple(pruned_skies_data.iloc[0][["Date", "Time"]])
sky_start_index = 0
skies_indexes = [] # (start_index, length)
for index, series in pruned_skies_data.iterrows():
    current_date_time = tuple(series[["Date", "Time"]])
    if prev_date_time != current_date_time:
        prev_date_time = current_date_time
        skies_indexes.append((sky_start_index, index - sky_start_index))
        sky_start_index = index
skies_indexes.append((sky_start_index, pruned_skies_data.shape[0] - sky_start_index))

sky_number_column = []
for number, index_and_length in enumerate(skies_indexes):
    sky_number_column += [number for _ in range(index_and_length[1])]

pruned_skies_data.insert(0, "SkyNumber", sky_number_column, False)

random.seed(0)
random.shuffle(skies_indexes)

indexes = []
for index, length in skies_indexes:
    indexes += [index + i for i in range(length)]

shuffled_skies = pruned_skies_data.iloc[indexes].reset_index(drop=True)
    
training_ratio = 0.75
sky_index = 0
prev_sky_number = int(shuffled_skies.iloc[0]["SkyNumber"])
n_skies = len(skies_indexes)
separator_index = 0
for index, series in shuffled_skies.iterrows():
    current_sky_number = int(series["SkyNumber"])
    if current_sky_number != prev_sky_number:
        prev_sky_number = current_sky_number
        sky_index += 1
        if (sky_index / n_skies) >= 0.75:
            separator_index = index
            break

train_skies_data = pruned_skies_data.iloc[:separator_index].sample(frac=1.0, random_state=0)
test_skies_data = pruned_skies_data.iloc[separator_index:].sample(frac=1.0, random_state=0)
```

```python
model = nn.Sequential(
    nn.Conv2d(3 * 8, 64, 7, padding=(3, 3)), # 100x100
    nn.MaxPool2d(2, 2), # 50x50
    nn.ReLU(inplace=True),

    nn.Conv2d(64, 128, 5, padding=(2, 2)), # 50x50
    nn.MaxPool2d(2, 2), # 25, 25
    nn.ReLU(inplace=True),
    
    nn.Conv2d(128, 256, 3, padding=(1, 1)), # 25x25
    nn.AdaptiveMaxPool2d(16), # 16x16
    nn.ReLU(inplace=True),
    
    nn.Conv2d(256, 512, 2, stride=(2, 2), padding=(0, 0)), # 8x8
    nn.ReLU(inplace=True),

    nn.Conv2d(512, 1024, 2, stride=(2, 2), padding=(0, 0)), # 4x4
    nn.ReLU(inplace=True),
    
    nn.Flatten(),

    nn.Dropout(),
    nn.Linear(1024 * 4 * 4, 8192),
    nn.ReLU(inplace=True),
    
    nn.Dropout(p=0.7),
    nn.Linear(8192, 4096),
    nn.ReLU(inplace=True),
    
    nn.Linear(4096, n_frequencies)
)

train_model = True
epochs = 100
batch_size = 100
```

```python
state_dict_save = None
if train_model:
    start_time = time.time()

    model.cuda()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.000001)
    criterion = torch.nn.MSELoss(reduction='sum')
    first_epochs_training_loss = 0
    first_epochs_testing_loss = 0
    
    prev_testing_loss = 0
    
    for epoch in range(epochs):
        total_training_loss = 0
        total_testing_loss = 0

        train_n_batches = math.ceil(train_skies_data.shape[0] / batch_size)
        model.train()
        for i in range(train_n_batches):
            start = i * batch_size
            samples_indexes = list(train_skies_data[start: start + batch_size]["TensorIndex"])

            batch = sky_sample_tensors[samples_indexes].cuda()
            ground_truth = normalized_y_tensor[samples_indexes].cuda()

            pred = model(batch)
            loss = criterion(pred, ground_truth)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_training_loss += float(loss)


        test_n_batches = math.ceil(test_skies_data.shape[0] / batch_size)
        model.eval()
        for i in range(test_n_batches):
            start = i * batch_size
            samples_indexes = list(test_skies_data[start: start + batch_size]["TensorIndex"])

            batch = sky_sample_tensors[samples_indexes].cuda()
            ground_truth = normalized_y_tensor[samples_indexes].cuda()

            pred = model(batch)
            loss = criterion(pred, ground_truth)

            total_testing_loss += float(loss)
            

        tab_char = '\t'
        print(f"{epoch}\tTraining Loss: {total_training_loss:.2f}\t{tab_char if total_training_loss < 100000 else ''}Testing Loss: {total_testing_loss:.2f}")
        if epoch == 0:
            first_epochs_training_loss = total_training_loss
            first_epochs_testing_loss = total_testing_loss
            prev_testing_loss = total_testing_loss
            state_dict_save = model.state_dict()

        change_in_training = total_training_loss - first_epochs_training_loss # I could use a func to calculate change
        change_in_training = change_in_training * 100 / first_epochs_training_loss
        change_in_training = " " * 15 + f'{"+" if change_in_training >= 0 else ""}{change_in_training:.2f}'

        change_in_testing = total_testing_loss - first_epochs_testing_loss
        change_in_testing = change_in_testing * 100 / first_epochs_testing_loss
        change_in_testing = " " * 22 + f'{"+" if change_in_testing >= 0 else ""}{change_in_testing:.2f}'

        print(f"\t{change_in_training}%\t{change_in_testing}%")

        minutes_elapsed_since_start = round((time.time() - start_time) / 60)
        print(f"Minutes Training: {minutes_elapsed_since_start}\n")
        
        if total_testing_loss < prev_testing_loss:
            state_dict_save = model.state_dict()
            
        prev_testing_loss = total_testing_loss

    model.load_state_dict(state_dict_save)
    model.cpu()
    torch.save(state_dict_save, '..\..\..\sky_model_sky_sampling')

else:
    model.load_state_dict(torch.load('model_dicts\sky_model_sky_sampling_500_epoc'))
    model.eval()
```

```python
test_tensor_indexes = list(test_skies_data["TensorIndex"])

truth = normalized_y_tensor[test_tensor_indexes]
truth_visible_range = truth[:,(380-350):(780-350)]
truth_infrared = truth[:,(780-350):]
truth.shape
```

```python
pred = model(sky_sample_tensors[test_tensor_indexes]).detach()
pred_visible_range = pred[:,(380-350):(780-350)]
pred_infrared = pred[:,(780-350):]
pred.shape
```

```python
range_all = y_range
range_visible = y_range[(380-350):(780-350)]
range_infrared = y_range[(780-350):]

def rmsd(pred_t, truth_t, range_t):
    diff_squared = ((pred_t - truth_t) * torch.tensor(range_t).float()) ** 2
    return diff_squared.sum(dim=1).mean().sqrt() * 100
    
rmsd_all = rmsd(pred, truth, range_all)
rmsd_visible = rmsd(pred_visible_range, truth_visible_range, range_visible)
rmsd_infrared = rmsd(pred_infrared, truth_infrared, range_infrared)
rmsd_all, rmsd_visible, rmsd_infrared
```
