```python
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
import time

from IPython.display import Image
import torch
from collections import defaultdict
import imageio
import numpy as np
from tqdm import tqdm

import torch
import torch.nn as nn

import math

from get_tensor import get_samples, get_batch, display_image
```

```python
csv_src = "..\..\..\coords-scattered-hdr.csv"
skies_data = pd.read_csv(csv_src)
# using a predefined random state to be able to save the samples
skies_data
```

```python
display_image(get_samples(skies_data.iloc[0]).cpu()[18:21])
```

```python
spectrum_beginning = 350
spectrum_end = 1780
n_frequencies = 1780 - 350 + 1

y = skies_data[
    [
        str(number)
        for number in range(spectrum_beginning, spectrum_end + 1)
    ]
]

y_mean = y.mean(0)
y_range = y.max(0) - y.min(0)
y_std = y.std(0)
normalized_y = (y - y_mean) / y_range # maybe use range instead of std
normalized_y_tensor = torch.tensor(normalized_y.values).float()
print(normalized_y_tensor.shape)
normalized_y
```

```python
# normalization
# we divide the brightness by 255 in the get_samples code
loading = True
save_location = '..\..\..\sky_samples_not_random'
if loading:
    sky_sample_tensors = torch.load('..\..\..\sky_samples_not_random')
else:
    sky_sample_tensors = get_batch(skies_data, 0, len(skies_data), get_samples)
    torch.save(sky_sample_tensors, save_location)

sky_sample_tensors.shape, sys.getsizeof(sky_sample_tensors.storage())
```

```python
# These skies are the ones we will use to validate our model against the original model
excluded_skies = [
    ("", "") # (date, time)
]
samples_per_sky = 81 # sample index 0-80


batch_size = 128
training_ratio = 0.75

n_samples = skies_data.shape[0]
n_batches = math.ceil(n_samples / batch_size)
testing_batches = math.ceil(training_ratio * n_batches)
```

```python
model = nn.Sequential(
    nn.Conv2d(3 * 8, 64, 7, padding=(3, 3)), # 100x100
    nn.BatchNorm2d(64),
    nn.MaxPool2d(2, 2), # 50x50
    nn.ReLU(inplace=True),

    nn.Conv2d(64, 128, 5, padding=(2, 2)), # 50x50
    nn.BatchNorm2d(128),
    nn.MaxPool2d(2, 2), # 25, 25
    nn.ReLU(inplace=True),
    
    nn.Conv2d(128, 256, 3, padding=(1, 1)), # 25x25
    nn.BatchNorm2d(256),
    nn.AdaptiveMaxPool2d(16), # 16x16
    nn.ReLU(inplace=True),
    
    nn.Conv2d(256, 512, 2, stride=(2, 2), padding=(0, 0)), # 8x8
    nn.BatchNorm2d(512),
    nn.ReLU(inplace=True),

    nn.Conv2d(512, 1024, 2, stride=(2, 2), padding=(0, 0)), # 4x4
    nn.BatchNorm2d(1024),
    nn.ReLU(inplace=True),
    
    nn.Flatten(),

    nn.Dropout(),
    nn.Linear(1024 * 4 * 4, 8192),
    nn.BatchNorm1d(8192),
    nn.ReLU(inplace=True),
    
    nn.Dropout(p=0.7),
    nn.Linear(8192, 4096),
    nn.BatchNorm1d(4096),
    nn.ReLU(inplace=True),
    
    nn.Linear(4096, n_frequencies)
)

model.cuda()
optimizer = torch.optim.Adam(model.parameters(), lr=0.000001)
criterion = torch.nn.MSELoss(reduction='sum')
model
```

```python
first_epochs_training_loss = 0
first_epochs_testing_loss = 0
for epoch in range(300):
    total_training_loss = 0
    total_testing_loss = 0

    for i in range(n_batches):
        if i == 0:
            model.train()
        if i == testing_batches:
            model.eval()

        start = i * batch_size
        batch = sky_sample_tensors[start: start + batch_size].cuda()
        ground_truth = normalized_y_tensor[start: start + batch_size].cuda()

        pred = model(batch)
        loss = criterion(pred, ground_truth)
        
        if i < testing_batches:
            # Training Batches Are Being used
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_training_loss += float(loss)
        else:
            # Testing Batches
            total_testing_loss += float(loss)

    print(f"{epoch}\tTraining Loss: {total_training_loss:.2f}\tTesting Loss: {total_testing_loss:.2f}")
    if epoch == 0:
        first_epochs_training_loss = total_training_loss
        first_epochs_testing_loss = total_testing_loss
    
    change_in_training = total_training_loss - first_epochs_training_loss # I could use a func to calculate change
    change_in_training = change_in_training * 100 / first_epochs_training_loss
    change_in_training = " " * 15 + f'{"+" if change_in_training >= 0 else ""}{change_in_training:.2f}'
    
    change_in_testing = total_testing_loss - first_epochs_testing_loss
    change_in_testing = change_in_testing * 100 / first_epochs_testing_loss
    change_in_testing = " " * 14 + f'{"+" if change_in_testing >= 0 else ""}{change_in_testing:.2f}'

    print(f"\t{change_in_training}%\t{change_in_testing}%\n")
    
torch.save(model.cpu().state_dict(), '..\..\..\sky_model_scattered')
```

```python
loaded_model = nn.Sequential(
    nn.Conv2d(3 * 8, 64, 7, padding=(3, 3)), # 100x100
    nn.BatchNorm2d(64),
    nn.MaxPool2d(2, 2), # 50x50
    nn.ReLU(inplace=True),

    nn.Conv2d(64, 128, 5, padding=(2, 2)), # 50x50
    nn.BatchNorm2d(128),
    nn.MaxPool2d(2, 2), # 25, 25
    nn.ReLU(inplace=True),
    
    nn.Conv2d(128, 256, 3, padding=(1, 1)), # 25x25
    nn.BatchNorm2d(256),
    nn.AdaptiveMaxPool2d(16), # 16x16
    nn.ReLU(inplace=True),
    
    nn.Conv2d(256, 512, 2, stride=(2, 2), padding=(0, 0)), # 8x8
    nn.BatchNorm2d(512),
    nn.ReLU(inplace=True),

    nn.Conv2d(512, 1024, 2, stride=(2, 2), padding=(0, 0)), # 4x4
    nn.BatchNorm2d(1024),
    nn.ReLU(inplace=True),
    
    nn.Flatten(),

    nn.Dropout(),
    nn.Linear(1024 * 4 * 4, 8192),
    nn.BatchNorm1d(8192),
    nn.ReLU(inplace=True),
    
    nn.Dropout(p=0.7),
    nn.Linear(8192, 4096),
    nn.BatchNorm1d(4096),
    nn.ReLU(inplace=True),
    
    nn.Linear(4096, n_frequencies)
)

loaded_model.load_state_dict(torch.load('..\..\..\sky_model_scattered'))
loaded_model.eval()
```

```python
pred = loaded_model(sky_sample_tensors[testing_batches * batch_size:])
truth = normalized_y_tensor[testing_batches * batch_size:]
```

```python
rmsd = (((pred - truth) * torch.tensor(y_range).float()) ** 2).mean(dim=1).sqrt().mean() * 100
rmsd
```
